# Exploratory Data Analysis


# Gender 

```python
df.columns
```

```python
gen=sns.countplot(x='Gender',data=df)

for bar in gen.containers:
    gen.bar_label(bar)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
```

```python
sales_gen
```

```python
k=sns.barplot(x='Gender',y='Amount',data=sales_gen)

```

From the above graph we can see that majority buyers are females. 

# Age

```python
df.columns
```

```python
age=sns.countplot(x='Age Group',data=df,hue='Gender')

for bar in age.containers:
    age.bar_label(bar)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the above graph we can see that majority buyers are from age between 26-35 and in that also majority are females.

# State

```python
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='State',y='Orders',data=sales_state)
sns.set(rc={'figure.figsize':(25,10)})
```

```python
sales_state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(x='State',y='Amount',data=sales_state_amount)
```

From the above graphs we can see that uttar pradesh has highest number of orders and also has highest sales amount value.

```python
sales_marital=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(1,5)})
```

```python
sales_marital_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Gender',ascending=False)
```

```python
sns.barplot(x='Marital_Status',y='Amount',data=sales_marital_amount,hue='Gender')
sns.set(rc={'figure.figsize':(7,5)})
```

From the above graph we can see that majority buyers are married female.

# Occupation

```python
df.columns
```

```python
sales_occu=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bar in sales_occu.containers:
    sales_occu.bar_label(bar)
    
```

```python
sales_occu_amount=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sns.barplot(x='Occupation',y='Amount',data=sales_occu_amount)
```

From the above graph we can see that majority orders are from IT sector and amount spend is also from IT sector

# Product Category

```python
df.columns
```

```python
sales_product=sns.countplot(x='Product_Category',data=df)
```

```python
sales_product_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
```
